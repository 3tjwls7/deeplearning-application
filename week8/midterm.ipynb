```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
# 데이터 로드
url = "https://raw.githubusercontent.com/MyungKyuYi/AI-class/main/kc_house_data.csv"
df = pd.read_csv(url)
df
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
# X(입력 변수), y(타겟 변수: price) 설정
X = df.drop('price', axis=1)
y = df['price']
```

```python
#원-핫 인코딩
Y = pd.get_dummies(y).values
```

```python
X = X.values
```

```python
X
```

```python
y
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape,
```

```python
#Defining the model

from keras.models import Sequential
from keras.layers import Dense,  Input
from keras.optimizers import SGD,Adam


model = Sequential()

model.add(Dense(10,input_shape=(20,),activation='relu'))
model.add(Dense(30,activation='relu'))
model.add(Dense(30,activation='relu'))
model.add(Dense(1))

model.compile(optimizer = 'rmsprop', loss = 'mse', metrics = ['mse'])

model.summary()
```

```python
N_EPOCHS=30
history=model.fit(x=X_train, y=y_train, epochs=N_EPOCHS, batch_size=32, validation_split= 0.1)
y_pred = model.predict(X_test)
```

```python
# loss 체크
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
plt.plot(range(N_EPOCHS), history.history['loss'], label='train loss')
plt.plot(range(N_EPOCHS), history.history['val_loss'], label='test loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
model.evaluate(X_test, y_test)
```
